## Creación de los gráficos

Se importan las siguientes librerías:

```python
import numpy as np                 #libreria para operaciones de algebra lineal
import pandas as pd                #librería para manejo de datos
import matplotlib.pyplot as plt    #librería para creación de gráficos
```

Supongamos que necesitamos un gráfico que muestre los ahorrros que hemos tenido durante los últimos 4 meses.

Declaramos los ahorros **en una lista:**

```python
ahorros = [1200, 400, 700, 1400, 2000, 660, 400, 980, 700, 1340, 560, 200]
```

Si quisieramos ver una gráfica de los ahorros, solo debemos usar el método *plot* que pertenece a la librería **matplotlib**

```python
plt.plot(ahorros)
```

## Usando un vector en lugar de una lista

```python
ahorros2 = np.array([1200, 400, 700, 1400, 2000, 660, 400, 980, 700, 1340, 560, 200])
plt.plot(ahorros2)
```

## Usando DataFrames

En estadística un **dataframe** hace alución a un marco de datos. Un **marco de datos** es una estructura básica para modelos estadísticos. Un ejemplo básico de un marco de datos puede ser una **hoja de cálculo**.

El Python la librería **Pandas** que sirve para editar, crear y manipular fácilmente un dataframe.

Definimos un dataframe usando al siguiente sintaxis:


```python
meses = ['Enero', 'Febrero', 'Marzo', 'Abril','Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
df = pd.DataFrame(ahorros, index = meses)
```

La línea de código anterior crea un dateframe llamado **df**, utilizando la libreria pandas (pd) y llamando al método *DataFrame* de la libreria mencionada. Se le pasa como argumento el vector (o la lista) de ahorros, así como una lista que contiene 12 *cadenas de caracteres* que corresponden a los nombres de los **meses**.

Gráficamos el dataframe usando *Matplotlib*:

```python
plt.title("Ahorros por mes")   #línea para poner nombre superior a la gráfica
plt.rc('font', size=6)              #línea es para cambiar el tamaño de las letras y números
plt.plot(df)
```
